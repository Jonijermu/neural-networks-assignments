```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Load GloVe embeddings
embeddings = {}
with open("../data/glove.6B.100d.txt", encoding="utf-8") as f:
    for line in f:
        values = line.split()
        word = values[0]
        vector = np.asarray(values[1:], dtype="float32")
        embeddings[word] = vector

def cosine_sim(a, b):
    return cosine_similarity(a.reshape(1, -1), b.reshape(1, -1))[0][0]

def find_closest(vec, embeddings, top_n=5):
    similarities = {}
    for word, emb in embeddings.items():
        similarities[word] = cosine_sim(vec, emb)
    return sorted(similarities.items(), key=lambda x: x[1], reverse=True)[:top_n]

# Words
man = embeddings["man"]
woman = embeddings["woman"]
king = embeddings["king"]

# Analogy
result_vec = woman - man + king

# Find nearest words
closest_words = find_closest(result_vec, embeddings)
print(closest_words)
```
