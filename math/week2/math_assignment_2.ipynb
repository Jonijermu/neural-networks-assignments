```python
def ReLU(x):
    return np.maximum(0, x)

def softmax(x):
    y = np.exp(x)
    return y / np.sum(y, axis=0)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
A0 = np.array([[3],
              [2],
              [5]])

W1 = np.array([[1.1, -1.3, 1.5],
              [0.1, 2.0, 1.7]])

W2 = np.array([[2.0, -1.2]])
```

```python
A1 = W1 @ A0
A2 = W2 @ A1

print(A1)
print(A2)
```

```python
x = np.arange(1,11).reshape(10, 1)
x
```

```python
y = np.array([14, 12, 34, 24, 43, 38, 41, 57, 55, 50]).reshape(10, 1)
y
```

```python
X = np.concatenate((np.ones((10, 1)), x), axis=1)
X
```

```python
xx = np.linspace(1, 10, 1000).reshape(1000, 1)
XX = np.concatenate((np.ones((1000, 1)), xx), axis=1)
M = la.inv(X.T @ X) @ X.T @ y
yy = XX.dot(M)
plt.plot(x, y, 'o')
plt.plot(xx, yy)
plt.show()
```
